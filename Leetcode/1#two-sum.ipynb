## Description
Given an array of integers nums and an integer target, return indices of the two numbers such that they add up to target.


You may assume that each input would have exactly one solution, and you may not use the same element twice.


You can return the answer in any order.


#### Example 1:
Input: nums = [2,7,11,15], target = 9
Output: [0,1]
Explanation: Because nums[0] + nums[1] == 9, we return [0, 1].


#### Example 2:
Input: nums = [3,2,4], target = 6
Output: [1,2]


#### Example 3:
Input: nums = [3,3], target = 6
Output: [0,1]

 

#### Constraints:

2 <= nums.length <= 104

-109 <= nums[i] <= 109

-109 <= target <= 109

Only one valid answer exists.
 

Follow-up: Can you come up with an algorithm that is less than O(n2) time complexity?

General Idea: Create a hashmap to store each element and its index in the nums array so that we can refer to the index once we find the counterpart.

- Steps:

1. Create a hashmap
2. Iterate through the array
    - For each number, calsulate its complement (target - current number);
    - If the complement exits in the hashmap, then return the indecies of the complement and the current number;
    - Otherwise, store the current number and its index.
3. The problem guarantees there's only one solution so we can return as soon as we find the match

### 4. **Implement**
```python
# Python implementation
class Solution:
    def twoSum(self, nums: List[int], target: int) -> List[int]:
        records = dict()
        for index, num in enumerate(nums):
            complement = target - num
            if complement in records:
                return [records[complement], index]
            records[num] = index

```

### 6. **Evaluate**
> **Time Complexity**: O(N), where N is the length of the array. We iterate through the array once, and each lookup in the hashmap is O(1).
> **Space Complexity**: O(N), where N is the length of the array. In the worst case, we store all the elements in the hashmap.
